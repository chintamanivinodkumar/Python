```python
from bs4 import BeautifulSoup
import requests
import csv
url = 'https://stats.espncricinfo.com/ci/engine/stats/index.html?class=3;result=1;template=results;type=batting'
p = requests.get(url)
soup = BeautifulSoup(p.content, 'html.parser')
 
content = soup.find_all('tr', class_ = 'data1')
Headrow =['player','span', 'mat','inns', 'battingDecesending','Hs', 'Bf','Sr','Ave',]
itemsfull=[]
 
for item in content:
  items =[]
  playerdata = item.find_all('td')
  for i in playerdata:
    items.append(i.text)
  itemsfull.append(items)
 
  
with open('T20Isbatting.csv','w') as csvfile:
  csvwriter = csv.writer(csvfile)
  csvwriter.writerow(Headrow)
  csvwriter.writerows(itemsfull)
 
 
  
import pandas as pd
data=pd.read_csv('T20Isbatting.csv')
data
```
