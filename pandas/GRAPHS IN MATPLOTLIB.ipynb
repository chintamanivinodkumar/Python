```python
#PLotting the data

import matplotlib.pyplot as plt
x = data["TripLength"]
y = data["MaxSpeed"]
plt.plot(x,y,color = "Blue")
plt.xlabel("TripLength")
plt.ylabel("MaxSpeed")
plt.title("Line plot")
plt.show()
```

```python
#Bar Graph

import matplotlib.pyplot as plt
x = data["TripLength"]
y = data["MaxSpeed"]
plt.bar(x,y,color = "Red")
plt.xlabel("TripLength")
plt.ylabel("MaxSpeed")
plt.title("Line plot")
plt.show()
```

```python
#scatter

import matplotlib.pyplot as plt
x = data["Honking"]
y = data["TripDuration"]
plt.scatter(x,y,color = "Black")
plt.xlabel("Honking")
plt.ylabel("TripDuration")
plt.title("Line plot")
plt.show()
```

```python
#pie chart

import matplotlib.pyplot as plt
x = data["Honking"]
plt.pie(x,shadow = True)
plt.ylabel("Honking")
plt.title("Pie plot")
plt.show()
```

```python
#Histogram plot

import matplotlib.pyplot as plt
x = data["Honking"]
y = data["Brakes"]
plt.hist(y,color = "Green")
plt.xlabel("Honking")
plt.ylabel("Brakes")
plt.title("Histogram")
plt.show()
```

```python
#Box plot

import matplotlib.pyplot as plt
items = [data.TripLength,data.TripDuration,data.MostFreqSpeed,data.Brakes]
plt.boxplot(items)
plt.title("Box Plot")
plt.show()
```
