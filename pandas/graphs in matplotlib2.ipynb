```python
import pandas as pd
df=pd.read_csv("nyc_weather.csv")
df
```

```python
#PLotting the data

import matplotlib.pyplot as plt
x = df["Temperature"]
y = df["DewPoint"]
plt.plot(x,y,color = "Blue")
plt.xlabel("TripLength")
plt.ylabel("MaxSpeed")
plt.title("Line plot")
plt.show()
```

```python
#Bar Graph

import matplotlib.pyplot as plt
x = df["Humidity"]
y = df["VisibilityMiles"]
plt.bar(x,y,color = "Red")
plt.xlabel("TripLength")
plt.ylabel("MaxSpeed")
plt.title("Line plot")
plt.show()
```

```python
#scatter

import matplotlib.pyplot as plt
x = df["PrecipitationIn"]
y = df["CloudCover"]
plt.scatter(x,y,color = "Black")
plt.xlabel("Honking")
plt.ylabel("TripDuration")
plt.title("Line plot")
plt.show()
```

```python
#pie chart

import matplotlib.pyplot as plt
x = df["Temperature"]
plt.pie(x, shadow = True)
plt.ylabel("DewPoint")
plt.title("Pie plot")
plt.show()
```

```python
#Histogram plot

import matplotlib.pyplot as plt
x = df["DewPoint"]
y = df["Humidity"]
plt.hist(y,color = "Green")
plt.xlabel("DewPoint")
plt.ylabel("Humidity")
plt.title("Histogram")
plt.show()
```
