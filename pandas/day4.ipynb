```python
import pandas as pd
df=pd.read_csv("weather_data.csv")
df
```

```python
df=pd.read_csv("weather_data.csv",parse_dates=['day'])
df
```

```python
df.set_index('day',inplace=True)
```

```python
df
```

```python
df.isna().sum()
```

```python
new_df=df.fillna(0) #filling null values with zero
new_df
```

```python
#filling the specific values to the particular column

new_df=df.fillna({
    'temperature' : 0,
    'windspeed' : 0,
    'event' : 'No event'
})
new_df
```

```python
new_df=df.fillna(method='ffill',limit=1) #forward filling method
new_df
```

```python
new_df=df.fillna(method='ffill',limit=2) #forward filling method
new_df
```

```python
new_df=df.fillna(method='bfill',limit=1) #backward filling method
new_df
```

```python
new_df=df.interpolate()
new_df
```

```python
new_df=df.interpolate(method='time')
new_df
```

```python
new_df=df.dropna()
new_df
```

```python
new_df=df.dropna(how='all') #how='all' drops the rows which consists all null values
new_df
```

```python
new_df=df.dropna(thresh=1) #keeping the row that has atleast one non null value
new_df
```

```python
new_df=df.dropna(thresh=2) #keeping the row that has two non null value
new_df
```
